# **Tic Tac Toe**

Tic-tac-toe, noughts and crosses, or Xs and Os is a paper-and-pencil game for two players who take turns marking the spaces in a three-by-three grid with X or O. The player who succeeds in placing three of their marks in a horizontal, vertical, or diagonal row is the winner.


## **Building the Grid**

In the below code chunk, 3*3 grid id built and saved into a function for reusability.

```python
startendline = ("|" + "-"*9)*3 + "|"
staticline = ("|" + " "*9)*3 + "|"
markupline1_1 = "|" + " "*9
markupline1_2 = "|" + " "*9
markupline1_3 = "|" + " "*9+'|'
markupline1 = markupline1_1 + markupline1_2 + markupline1_3
markupline2_1 = "|" + " "*9
markupline2_2 = "|" + " "*9
markupline2_3 = "|" + " "*9+'|'
markupline2 = markupline2_1 + markupline2_2 + markupline2_3
markupline3_1 = "|" + " "*9
markupline3_2 = "|" + " "*9
markupline3_3 = "|" + " "*9+'|'
markupline3 = markupline3_1 + markupline3_2 + markupline3_3

def default_board():
  print(startendline)
  print(staticline)
  print(markupline1)
  print(staticline)
  print(startendline)
  print(staticline)
  print(markupline2)
  print(staticline)
  print(startendline)
  print(staticline)
  print(markupline3)
  print(staticline)
  print(startendline)
```

## **Restarting the game**

If the game is over and the player wants to player one more game the grid is reset and calls the tictactoe again to play new game.

```python
def replay():
  global markupline1
  global markupline2
  global markupline3
  global markupline1_1
  global markupline1_2
  global markupline1_3
  global markupline2_1
  global markupline2_2
  global markupline2_3
  global markupline3_1
  global markupline3_2
  global markupline3_3
  global startendline
  global staticline
  startendline = ("|" + "-"*9)*3 + "|"
  staticline = ("|" + " "*9)*3 + "|"
  markupline1_1 = "|" + " "*9
  markupline1_2 = "|" + " "*9
  markupline1_3 = "|" + " "*9+'|'
  markupline1 = markupline1_1 + markupline1_2 + markupline1_3
  markupline2_1 = "|" + " "*9
  markupline2_2 = "|" + " "*9
  markupline2_3 = "|" + " "*9+'|'
  markupline2 = markupline2_1 + markupline2_2 + markupline2_3
  markupline3_1 = "|" + " "*9
  markupline3_2 = "|" + " "*9
  markupline3_3 = "|" + " "*9+'|'
  markupline3 = markupline3_1 + markupline3_2 + markupline3_3
  tictactoe()
```

## **Allocating the cards**

Here in the function board_position, the cards(crosses and noughts) are placed as per the players response and the grid will updated with card at the chosen position and return the board for every response.

```python
def board():
  global markupline1
  global markupline2
  global markupline3
  global markupline1_1
  global markupline1_2
  global markupline1_3
  global markupline2_1
  global markupline2_2
  global markupline2_3
  global markupline3_1
  global markupline3_2
  global markupline3_3
  global startendline
  global staticline
  print(startendline)
  print(staticline)
  print(markupline1_1+markupline1_2+markupline1_3)
  print(staticline)
  print(startendline)
  print(staticline)
  print(markupline2_1+markupline2_2+markupline2_3)
  print(staticline)
  print(startendline)
  print(staticline)
  print(markupline3_1+markupline3_2+markupline3_3)
  print(staticline)
  print(startendline)
```

```python
def board_position(position, card):
  global markupline1
  global markupline2
  global markupline3
  global markupline1_1
  global markupline1_2
  global markupline1_3
  global markupline2_1
  global markupline2_2
  global markupline2_3
  global markupline3_1
  global markupline3_2
  global markupline3_3
  if position == 1:
    markupline1_1 = '|    ' + card + '    '
  elif position == 2:
    markupline1_2 = '|    ' + card + '    '
  elif position == 3:
    markupline1_3 = '|    ' + card + '    |'
  elif position == 4:
    markupline2_1 = '|    ' + card + '    '
  elif position == 5:
    markupline2_2 = '|    ' + card + '    '
  elif position == 6:
    markupline2_3 = '|    ' + card + '    |'
  elif position == 7:
    markupline3_1 = '|    ' + card + '    '
  elif position == 8:
    markupline3_2 = '|    ' + card + '    '
  elif position == 9:
    markupline3_3 = '|    ' + card + '    |'
  return board()
```

## The Game

The tictactoe is the function where all the game takes place.

1. input function is used to get the players names and their chosen cards (cross or nought).
2. random.choice to pick a player to play first randomly.
3. The gridcounter makes sure not more than 9 plays
4. positionlist makes sure the grid that is filled with card already not be used or filled.
5. cardlist helps us to find the winner according to last added card to the list.
6. trigger variable is set to True to process the game and is set to False if there is a win.
7. Once the game is done we make use of replay function to reset the grid and trigger the game again if the player response is yes else it will exit the game.

```python
def tictactoe():
  import random
  global markupline1
  global markupline2
  global markupline3
  global markupline1_1
  global markupline1_2
  global markupline1_3
  global markupline2_1
  global markupline2_2
  global markupline2_3
  global markupline3_1
  global markupline3_2
  global markupline3_3
  print("Welcome to the game!")
  P1 = input("Player1, Enter you name: ").capitalize()
  P2 = input("Player2, Enter your name: ").capitalize()
  print("Let's start the game!")
  playerlist = [P1,P2]
  print(input("Press Enter for toss."))
  firstplay = random.choice(playerlist)
  if firstplay == P1:
    print(f'{P1}, plays first.')
  else:
    P2 = P1
    P1 = firstplay
    print(f'{P1}, plays first.')
  while True:
    card1 = input(f"{P1}, Enter your sign X or O: ").upper()
    if card1 == 'X':
      card2 = 'O'
      break
    elif card1 =='O':
      card2 = 'X'
      break
    else:
      print("Invalid Sign, choose only 'X' or 'O'.")
      continue
  print(f'{P1}:{card1}, {P2}:{card2}')
  default_board()
  gridcounter = 0
  positionlist = []
  cardlist = []
  trigger = True
  while trigger:
    while gridcounter < 9:
        if gridcounter%2 == 0 and (not((markupline1_1[5] == markupline2_1[5] == markupline3_1[5]) and (markupline1_1[5] !=' ' or markupline2_1[5] != ' ' or markupline3_1[5] != ' ')) and not((markupline1_2[5] == markupline2_2[5] == markupline3_2[5]) and (markupline1_2[5] !=' ' or markupline2_2[5] != ' ' or markupline3_2[5] != ' ')) and not((markupline1_3[5] == markupline2_3[5] == markupline3_3[5]) and (markupline1_3[5] !=' ' or markupline2_3[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_1[5] == markupline1_2[5] == markupline1_3[5]) and (markupline1_1[5] !=' ' or markupline1_2[5] != ' ' or markupline1_3[5] != ' ')) and not((markupline2_1[5] == markupline2_2[5] == markupline2_3[5]) and (markupline2_1[5] !=' ' or markupline2_2[5] != ' ' or markupline2_3[5] != ' ')) and not((markupline3_1[5] == markupline3_2[5] == markupline3_3[5]) and (markupline3_1[5] !=' ' or markupline3_2[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_1[5] == markupline2_2[5] == markupline3_3[5]) and (markupline1_1[5] !=' ' or markupline2_2[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_3[5] == markupline2_2[5] == markupline3_1[5]) and (markupline1_3[5] !=' ' or markupline2_2[5] != ' ' or markupline3_1[5] != ' '))):
          place1 = int(input(f'{P1}, chose a number between 1-9: '))
          if place1 in positionlist or place1 not in range(1,10):
            print("Grid already filled or chosen number out of range.")
            continue
          else:
            positionlist.append(place1)
            cardlist.append(card1)
            gridcounter += 1
            board_position(place1,card1)
        elif gridcounter%2 != 0 and (not((markupline1_1[5] == markupline2_1[5] == markupline3_1[5]) and (markupline1_1[5] !=' ' or markupline2_1[5] != ' ' or markupline3_1[5] != ' ')) and not((markupline1_2[5] == markupline2_2[5] == markupline3_2[5]) and (markupline1_2[5] !=' ' or markupline2_2[5] != ' ' or markupline3_2[5] != ' ')) and not((markupline1_3[5] == markupline2_3[5] == markupline3_3[5]) and (markupline1_3[5] !=' ' or markupline2_3[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_1[5] == markupline1_2[5] == markupline1_3[5]) and (markupline1_1[5] !=' ' or markupline1_2[5] != ' ' or markupline1_3[5] != ' ')) and not((markupline2_1[5] == markupline2_2[5] == markupline2_3[5]) and (markupline2_1[5] !=' ' or markupline2_2[5] != ' ' or markupline2_3[5] != ' ')) and not((markupline3_1[5] == markupline3_2[5] == markupline3_3[5]) and (markupline3_1[5] !=' ' or markupline3_2[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_1[5] == markupline2_2[5] == markupline3_3[5]) and (markupline1_1[5] !=' ' or markupline2_2[5] != ' ' or markupline3_3[5] != ' ')) and not((markupline1_3[5] == markupline2_2[5] == markupline3_1[5]) and (markupline1_3[5] !=' ' or markupline2_2[5] != ' ' or markupline3_1[5] != ' '))):
          place1 = int(input(f'{P2}, chose a number between 1-9: '))
          if place1 in positionlist or place1 not in range(1,10):
            print("Grid already filled or chosen number out of range.")
            continue
          else:
            positionlist.append(place1)
            cardlist.append(card2)
            gridcounter += 1
            board_position(place1,card2)
        else:
          if cardlist.pop() == card1:
            print(f'{P1} wins!')
            gridcounter += 1
            trigger = False
          else:
            print(f'{P2} wins!')
            gridcounter += 1
            trigger = False
        break

    else:
      while (gridcounter > 8 and gridcounter < 10):
        if (markupline1_1[5] != markupline1_2[5] or markupline1_2[5] != markupline1_3[5] or markupline1_3[5] !=  markupline1_1[5]) and (markupline2_1[5] != markupline2_2[5] or markupline2_2[5] != markupline2_3[5] or markupline2_3[5] !=  markupline2_1[5]) and (markupline3_1[5] != markupline3_2[5] or markupline3_2[5] != markupline3_3[5] or markupline3_3[5] !=  markupline3_1[5]) and (markupline1_1[5] != markupline2_1[5] or markupline2_1[5] != markupline3_1[5] or markupline3_1[5] !=  markupline1_1[5]) and (markupline1_2[5] != markupline2_2[5] or markupline2_2[5] != markupline3_2[5] or markupline3_2[5] !=  markupline1_2[5]) and (markupline1_3[5] != markupline2_3[5] or markupline2_3[5] != markupline3_3[5] or markupline3_3[5] !=  markupline1_3[5]) and (markupline1_1[5] != markupline2_2[5] or markupline2_2[5] != markupline3_3[5] or markupline3_3[5] !=  markupline1_1[5]) and (markupline1_3[5] != markupline2_2[5] or markupline2_2[5] != markupline3_1[5] or markupline3_1[5] !=  markupline1_3[5]) :
          print("Match Draw")
          gridcounter += 1
        else:
          if cardlist.pop() == card1:
            print(f'{P1} wins!')
            gridcounter += 1
          else:
            print(f'{P2} wins!')
            gridcounter += 1
      break
  restart = input("Do you want to play one more game 'Yes' or 'No': ")
  if restart.upper() == 'YES':
    replay()
  else:
    print("Thank you for playing TIC TAC TOE!")
```
